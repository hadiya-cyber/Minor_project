# Project 1:
## Console-Based Smart Calculator: 
#This is a smart calculator which uses menu driven programs. It takes numbers as input from users and 
#calculates the numbers using operations which are also the users choice.

```python
import numpy as np

def calculator():
 while True:
  print("\n1.Add  2.Sub  3.Mul  4.Div  5.Power  6.Mod  7.Exit")
  ch=int(input("Choice: "))

  if ch==7:
   break

  n=int(input("How many numbers: "))
  a=[]

  for i in range(n):
   a.append(float(input("Enter number: ")))

  arr=np.array(a)

  if ch==1:
   print("Result:",np.sum(arr))

  elif ch==2:
   res=arr[0]
   for i in range(1,n):
    res-=arr[i]
   print("Result:",res)

  elif ch==3:
   res=1
   for i in arr:
    res*=i
   print("Result:",res)

  elif ch==4:
   if arr[1]==0:
    print("Division by zero")
   else:
    print("Result:",arr[0]/arr[1])

  elif ch==5:
   print("Result:",arr[0]**arr[1])

  elif ch==6:
   print("Result:",arr[0]%arr[1])

calculator()
```

# Project 2
## Student Marks Analyzer

```python
import numpy as np

def grade(avg):
 if avg>85:
  return "A"
 elif avg>=70:
  return "B"
 elif avg>=50:
  return "C"
 else:
  return "Fail"

n=int(input("Number of students: "))
m=[]

for i in range(n):
 m.append(float(input("Enter marks: ")))

arr=np.array(m)

print("Total:",np.sum(arr))
print("Average:",np.mean(arr))
print("Highest:",np.max(arr))
print("Lowest:",np.min(arr))
print("Grade:",grade(np.mean(arr)))
```

## Daily Expense Tracker

```python
# Problem Statement: The daily expenses of ann individual are taken and information is provided: Like total,average and highest amount spent 
# by individual based on calculations.
```

```python
import numpy as np

n=int(input("Number of days: "))
e=[]

for i in range(n):
 e.append(float(input("Enter expense: ")))

arr=np.array(e)

print("Total:",np.sum(arr))
print("Average:",np.mean(arr))
print("Highest:",np.max(arr))

if np.mean(arr)>500:
 print("Warning: Limit exceeded")
```
